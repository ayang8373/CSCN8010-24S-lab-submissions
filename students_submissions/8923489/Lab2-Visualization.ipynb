### 1. Heatmap visualization 
Explore correlations between features 


```python
plt.figure(figsize=(14,12))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### 2.CountPlot Visualization
Displays the frequency of age within a categorical variable

```python
plt.figure(figsize=(15,10))
sns.countplot(x=df['target'],hue=df['age'])
```

###  3. Scatterplot visualization
Correlations between features by plotting data points for age and index and their relationship.

```python
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(10, 4))
    sns.scatterplot(x=df.index, y=df[column], alpha=0.5, color='blue')
    
    plt.title(f'Scatter Plot of {column}')
    plt.xlabel('Index')
    plt.ylabel(column)
    plt.show()
```

### 4. Histogram Visualization
Represents the distribution of ages by depicting the frequency of occurrence of different age ranges within a dataset

```python
selected_columns = df.select_dtypes(include=['float64', 'int64']).columns

for column in selected_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
    break  
```

### 5.Boxplot Visualization
Boxplots visualize the distribution, median, and outliers of data

```python
selected_columns = df.select_dtypes(include=['float64', 'int64']).columns
    
for column in selected_columns:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=df[column])
        plt.title(f'Box Plot of {column}')
        plt.xlabel(column)
        plt.show()
        break
```

### 6. Pairplot
Grid of scatterplots where each variable in the DataFrame (df) is plotted against every other.

```python
sns.pairplot(df,hue='target')
```
