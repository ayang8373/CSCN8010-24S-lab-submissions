```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

### First chart is example of linear chart
data is just example not real data

```python
x = [100, 200, 300, 400, 500]
y = [10, 11, 12, 10, 10]

plt.figure()
plt.plot(x, y, marker='o')
plt.title("Average customer profit - (Basic chart)")
plt.xlabel("num of customers in store")
plt.ylabel("profit")
plt.grid(True)
plt.show()
```

### Most populated countries

### Mobile phone manufacturers
by market share

[data source](https://www.linkedin.com/pulse/top-10-most-selling-phone-world-so-far2024-voke-me-valmf/)

```python
labels = ['Apple', 'Samsung','Xiaomi','Oppo', 'Vivo']
sizes = [27.95, 23.95, 11.44, 5.77, 5]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'orange']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice

# Plot
plt.figure(figsize=(10, 5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Market share')
plt.axis('equal')
plt.show()
```

### GDP vs Life Expectancy
this is scatter plot displaying dependency between GDP p.c. and life expectancy

[data source](https://ourworldindata.org/grapher/life-expectancy-vs-gdp-per-capita?tab=table)

```python
data_le = {
    "Country": ["Canada", "Czech republic", "UK", "Zimbabwe", "World AVG."],
    "GDP": [44.380, 31.453, 37.134, 1.687, 16.283],  # in million USD
    "Life Expectancy": [82.7 , 77.7, 80.7, 59.3, 71]  # in years
}
plt.figure()
plt.scatter(data_le['GDP'], data_le['Life Expectancy'], color='red')
for i, txt in enumerate(data_le['Country']):
    plt.annotate(txt, (data_le['GDP'][i], data_le['Life Expectancy'][i]))
plt.title("GDP vs. Life Expectancy")
plt.xlabel("GDP per capacita(in thousand USD)")
plt.ylabel("Life Expectancy (years)")
plt.show()
```
