# Univariate Linear Regression on the Scikit-Learn Diabetes Dataset

my solution to linear regression using Scikit-Learn

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

diabetes_X_bmi = diabetes_X[:, np.newaxis, 2]

X_train, X_test, y_train, y_test = train_test_split(diabetes_X_bmi, diabetes_y, test_size=0.2, random_state=5)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)
mae_train = mean_absolute_error(y_train, y_pred_train)
mae_test = mean_absolute_error(y_test, y_pred_test)

print("Model slope:", round(model.coef_[0],3))
print("Model intercept:", round(model.intercept_,3))
print("Training MSE:", round(mse_train,3))
print("Test MSE:", round(mse_test,3))
print("Training MAE:", round(mae_train,3))
print("Test MAE:", round(mae_test,3))
```


Relatively high errors in MSE and MAE indicate that the data may be influenced by factors other than just BMI.

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

axs[0].scatter(X_train, y_train, color='blue', label='Training data')
axs[0].plot(X_train, y_pred_train, color='red', linewidth=2, label='Model')
axs[0].set_xlabel('BMI')
axs[0].set_ylabel('Disease Progression')
axs[0].set_title('Training Data: BMI vs Disease Progression')
axs[0].legend(loc='best')

axs[1].scatter(X_test, y_test, color='green', label='Testing data')
axs[1].plot(X_test, y_pred_test, color='red', linewidth=2, label='Model')
axs[1].set_xlabel('BMI')
axs[1].set_ylabel('Disease Progression')
axs[1].set_title('Testing Data: BMI vs Disease Progression')
axs[1].legend(loc='best')

plt.tight_layout()
plt.show()
```

Graphs displaying data and trends of linear regression and testing data.
